```python
# 1.Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`. After generating the list, find the following:

import numpy as np
from scipy import stats
from math import prod

# Generate the list of 100 integers between 90 and 130
int_list = np.random.randint(90, 131, size=100)

# (i) Mean and Median
def calculate_mean(lst):
    return np.mean(lst)

def calculate_median(lst):
    return np.median(lst)

# (ii) Mode
def calculate_mode(lst):
    mode, _ = stats.mode(lst)
    return mode[0]

# (iii) Weighted Mean
def weighted_mean(values, weights):
    return np.average(values, weights=weights)

# (iv) Geometric Mean
def geometric_mean(lst):
    return stats.gmean(lst)

# (v) Harmonic Mean
def harmonic_mean(lst):
    return stats.hmean(lst)

# (vi) Midrange
def midrange(lst):
    return (np.min(lst) + np.max(lst)) / 2

# (vii) Trimmed Mean
def trimmed_mean(lst, proportion_to_cut):
    return stats.trim_mean(lst, proportion_to_cut)

# Sample weights for weighted mean
weights = np.random.rand(100)

print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Weighted Mean:", weighted_mean(int_list, weights))
print("Geometric Mean:", geometric_mean(int_list))
print("Harmonic Mean:", harmonic_mean(int_list))
print("Midrange:", midrange(int_list))
print("Trimmed Mean:", trimmed_mean(int_list, proportion_to_cut=0.1))
```

```python
#3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance
import numpy as np

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        self.values = np.array(values)
        self.probabilities = np.array(probabilities)
        if not np.isclose(np.sum(self.probabilities), 1):
            raise ValueError("Probabilities must sum to 1")

    def expected_value(self):
        return np.sum(self.values * self.probabilities)

    def variance(self):
        mean = self.expected_value()
        return np.sum(self.probabilities * (self.values - mean) ** 2)

# Example usage
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]
rv = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
```

```python
# 5.Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance

import numpy as np

def generate_samples(dist_type, size=1000, **params):
    if dist_type == 'binomial':
        n = params.get('n')
        p = params.get('p')
        if n is None or p is None:
            raise ValueError("Parameters 'n' and 'p' are required for binomial distribution")
        samples = np.random.binomial(n=n, p=p, size=size)
    elif dist_type == 'poisson':
        lam = params.get('lam')
        if lam is None:
            raise ValueError("Parameter 'lam' is required for Poisson distribution")
        samples = np.random.poisson(lam=lam, size=size)
    else:
        raise ValueError("Unsupported distribution type")
    
    mean = np.mean(samples)
    variance = np.var(samples)
    return mean, variance

# Example usage
print("Binomial Distribution:", generate_samples('binomial', size=1000, n=10, p=0.5))
print("Poisson Distribution:", generate_samples('poisson', size=1000, lam=5))
```

```python
# 6. Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples
import numpy as np

def generate_gaussian_samples(mean, std_dev, size=1000):
    samples = np.random.normal(mean, std_dev, size=size)
    mean = np.mean(samples)
    variance = np.var(samples)
    std_dev = np.std(samples)
    return mean, variance, std_dev

# Example usage
print("Gaussian Distribution:", generate_gaussian_samples(mean=0, std_dev=1))
```

```python
#11.A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.
import numpy as np
from scipy import stats

# Data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Conversion rates
p_old = np.mean(old_layout)
p_new = np.mean(new_layout)

# Pooled proportion
p_pooled = (np.sum(old_layout) + np.sum(new_layout)) / (len(old_layout) + len(new_layout))

# Z-score and p-value calculation
z_score = (p_new - p_old) / np.sqrt(p_pooled * (1 - p_pooled) * (1 / len(old_layout) + 1 / len(new_layout)))
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))

print("Z-score for website layouts:", z_score)
print("P-value for website layouts:", p_value)
```

```python
#15.A company is testing two different website layouts to see which one leads to higher click-through rates. Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value. 
import numpy as np
from scipy import stats

def ab_test(layout_a_clicks, layout_b_clicks):
    # Data
    a = np.array(layout_a_clicks)
    b = np.array(layout_b_clicks)

    # Means
    mean_a = np.mean(a)
    mean_b = np.mean(b)

    # Variances
    var_a = np.var(a, ddof=1)
    var_b = np.var(b, ddof=1)

    # Sample sizes
    n_a = len(a)
    n_b = len(b)

    # T-statistic and degrees of freedom
    pooled_var = ((n_a - 1) * var_a + (n_b - 1) * var_b) / (n_a + n_b - 2)
    t_statistic = (mean_a - mean_b) / np.sqrt(pooled_var * (1 / n_a + 1 / n_b))
    df = n_a + n_b - 2

    # P-value calculation
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))

    return t_statistic, df, p_value

# Example usage
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
t_statistic, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

print("T-statistic for A/B test:", t_statistic)
print("Degrees of freedom:", df)
print("P-value for A/B test:", p_value)
```

```python
#16.A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the t-statistic and p-value for the treatment effect. 
import numpy as np
from scipy import stats

def t_test(existing_drug_levels, new_drug_levels):
    # Data
    existing = np.array(existing_drug_levels)
    new = np.array(new_drug_levels)

    # Means
    mean_existing = np.mean(existing)
    mean_new = np.mean(new)

    # Variances
    var_existing = np.var(existing, ddof=1)
    var_new = np.var(new, ddof=1)

    # Sample sizes
    n_existing = len(existing)
    n_new = len(new)

    # T-statistic and degrees of freedom
    pooled_var = ((n_existing - 1) * var_existing + (n_new - 1) * var_new) / (n_existing + n_new - 2)
    t_statistic = (mean_existing - mean_new) / np.sqrt(pooled_var * (1 / n_existing + 1 / n_new))
    df = n_existing + n_new - 2

    # P-value calculation
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))

    return t_statistic, df, p_value

# Example usage
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]
t_statistic, df, p_value = t_test(existing_drug_levels, new_drug_levels)

print("T-statistic for drug effect on cholesterol levels:", t_statistic)
print("Degrees of freedom:", df)
print("P-value for drug effect on cholesterol levels:", p_value)
```

```python
#17.A school district introduces an educational intervention program to improve math scores. Write a Python function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to determine if the intervention had a significant impact. 
import numpy as np
from scipy import stats

def t_test_pre_post(pre_intervention_scores, post_intervention_scores):
    # Data
    pre = np.array(pre_intervention_scores)
    post = np.array(post_intervention_scores)

    # Differences
    differences = post - pre

    # Mean and standard deviation of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    n = len(differences)

    # T-statistic
    t_statistic = mean_diff / (std_diff / np.sqrt(n))
    df = n - 1

    # P-value calculation
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))

    return t_statistic, df, p_value

# Example usage
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
t_statistic, df, p_value = t_test_pre_post(pre_intervention_scores, post_intervention_scores)

print("T-statistic for educational intervention:", t_statistic)
print("Degrees of freedom:", df)
print("P-value for educational intervention:", p_value)
```

```python
#18.An HR department wants to investigate if there's a gender-based salary gap within the company. Develop a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically significant difference between the average salaries of male and female employees. 
import numpy as np
from scipy import stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def t_test_two_samples(sample1, sample2):
    # Data
    sample1 = np.array(sample1)
    sample2 = np.array(sample2)

    # Means
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)

    # Variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Sample sizes
    n1 = len(sample1)
    n2 = len(sample2)

    # T-statistic and degrees of freedom
    pooled_var = ((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2)
    t_statistic = (mean1 - mean2) / np.sqrt(pooled_var * (1 / n1 + 1 / n2))
    df = n1 + n2 - 2

    # P-value calculation
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))

    return t_statistic, df, p_value

# Example usage
t_statistic, df, p_value = t_test_two_samples(male_salaries, female_salaries)
print("T-statistic for gender-based salary gap:", t_statistic)
print("Degrees of freedom:", df)
print("P-value for gender-based salary gap:", p_value)
```

```python
#21.A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences. 
import numpy as np
from scipy import stats

# Generate synthetic data
np.random.seed(0)  # For reproducibility
age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create contingency table
contingency_table = np.array([
    [np.sum((age_groups == '18-30') & (voter_preferences == 'Candidate A')),
     np.sum((age_groups == '18-30') & (voter_preferences == 'Candidate B'))],
    [np.sum((age_groups == '31-50') & (voter_preferences == 'Candidate A')),
     np.sum((age_groups == '31-50') & (voter_preferences == 'Candidate B'))],
    [np.sum((age_groups == '51+') & (voter_preferences == 'Candidate A')),
     np.sum((age_groups == '51+') & (voter_preferences == 'Candidate B'))]
])

# Perform Chi-Square test
chi2_statistic, p_value, _, _ = stats.chi2_contingency(contingency_table)

print("Chi-Square statistic:", chi2_statistic)
print("P-value:", p_value)
```
